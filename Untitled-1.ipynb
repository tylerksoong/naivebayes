```python
deez_nuts = {}
with open('SPAMTrain.label', 'r') as f:
    for line in f.readlines():
        splitted = line.split()
        deez_nuts[splitted[1]] = splitted[0]
```

```python
from email.parser import BytesParser
from email import policy
import os
from tqdm import tqdm
from collections import Counter

spam = Counter()
regular = Counter()
email_path = "TRAINING/"


for file_name in tqdm(os.listdir(email_path)):
    label = deez_nuts[file_name]
    try:
        with open(os.path.join(email_path, file_name), 'rb') as fp:
            msg = BytesParser(policy=policy.default).parse(fp)

        body = msg.get_body(preferencelist=('plain'))
        if body:
            words = body.get_content().split()
            if label == '0':
                regular.update(words)
            else:
                spam.update(words)
    except Exception as e:
        print(f"Skipping {file_name}: {e}")


```

```python




import re

def is_meaningful(word):
    return word.isalpha() and len(word) > 2  # letters only, no symbols/short tokens

spam_filtered = Counter({w: c for w, c in spam.items() if is_meaningful(w)})
regular_filtered = Counter({w: c for w, c in regular.items() if is_meaningful(w)})

# Get total counts for normalization
spam_total = sum(spam_filtered.values())
regular_total = sum(regular_filtered.values())


spam_top = dict(spam_filtered.most_common(1000))
regular_top = dict(regular_filtered.most_common(1000))


# Get all unique words from both top 100 lists
all_words = set(spam_top) | set(regular_top)

# Compute squared difference in proportions
squared_diffs = {
    word: abs(spam_top.get(word, 0) / spam_total - regular_top.get(word, 0) / regular_total)
    for word in all_words
}

# Sort by largest squared difference
sorted_diffs = sorted(squared_diffs.items(), key=lambda x: x[1], reverse=True)
```

```python
words = [x[0] for x in sorted_diffs[:800]]
words
print(len(sorted_diffs))
```

```python
from naive_bayes import create_matrix

df = create_matrix(training_dir='TRAINING/', features=words)
```

```python
import numpy as np
from naive_bayes import NaiveBayesBernoulli

nb = NaiveBayesBernoulli()
nrow, ncol = df.shape

X_train = df.iloc[:3500, 1:ncol-1]
y_train = df.iloc[:3500, ncol-1]
X_test = df.iloc[3500:, 1:ncol-1]
y_test = df.iloc[3500:, ncol-1]

nb.fit(X_train, y_train)

preds = [nb.predict(X_test.iloc[i])[0] for i in range(len(X_test))]
np.mean(preds == y_test.astype(int))
```
